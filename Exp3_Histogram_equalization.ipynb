```python
# import Opencv
import cv2

# import Numpy
import numpy as np

# read a image using imread
img = cv2.imread('/content/rabbit.png', 0)

# creating a Histograms Equalization
# of a image using cv2.equalizeHist()
equ = cv2.equalizeHist(img)

# stacking images side-by-side
res = np.hstack((img, equ))

# import cv2_imshow from google.colab.patches for image display in Colab
from google.colab.patches import cv2_imshow

# show image input vs output using cv2_imshow instead of cv2.imshow
cv2_imshow(res)

# cv2.waitKey(0) # Not needed in Colab when using cv2_imshow
# cv2.destroyAllWindows() # Not needed in Colab when using cv2_imshow
# cv2.destroyAllWindows() # Not needed in Colab when using cv2_imshow
```
